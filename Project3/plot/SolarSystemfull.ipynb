```python
    import numpy as np
    import matplotlib.pyplot as plt
    import pandas as pd
    from mpl_toolkits.axes_grid.axislines import SubplotZero
    from matplotlib.ticker import MultipleLocator, FormatStrFormatter
    fig = plt.figure(figsize=(5, 5))
    ax = SubplotZero(fig, 111)
    fig.add_subplot(ax)
    ax.grid(False)
    
    # Major and Minor Ticks setting
#     xmajorLocator   = MultipleLocator(1)
#     xmajorFormatter = FormatStrFormatter('%5.1f')
#     xminorLocator = MultipleLocator(0.5) 
#     ax.xaxis.set_major_locator(xmajorLocator)
#     ax.xaxis.set_major_formatter(xmajorFormatter)
#     ax.xaxis.set_minor_locator(xminorLocator)
#     ax.yaxis.set_major_locator(xmajorLocator)
#     ax.yaxis.set_major_formatter(xmajorFormatter)
#     ax.yaxis.set_minor_locator(xminorLocator)


    for direction in ["xzero", "yzero"]:
        ax.axis[direction].set_axisline_style("-|>")
        ax.axis[direction].set_visible(True)
        ax.axis[direction].line.set_color('k')
        ax.set_ylim(-1.8, 2.2)
        ax.set_xlim(-1.8, 2.2)
        ax.set_aspect(1)
       
    ax.axis["yzero"].set_ticklabel_direction("-")
    
    for direction in ["left", "right", "bottom", "top"]:
        ax.axis[direction].set_visible(False)

    sun = pd.read_csv("sun.csv")
    mercury = pd.read_csv("mercury.csv")
    venus= pd.read_csv("venus.csv")
    earth = pd.read_csv("earth.csv")
    mars = pd.read_csv("mars.csv")
    jupiter = pd.read_csv("jupiter.csv")
    saturn = pd.read_csv("saturn.csv")
    uranus = pd.read_csv("uranus.csv")
    neptune = pd.read_csv("neptune.csv")

 
    x1 = sun.iloc[:,3]; y1 = sun.iloc[:,4]
    x2 = mercury.iloc[:,3]; y2 = mercury.iloc[:,4]
    x3 = venus.iloc[:,3]; y3 = venus.iloc[:,4]
    x4 = earth.iloc[:,3]; y4 = earth.iloc[:,4]
    x5 = mars.iloc[:,3]; y5 = mars.iloc[:,4]
    x6 = jupiter.iloc[:,3]; y6 = jupiter.iloc[:,4]
    x7 = saturn.iloc[:,3]; y7 = saturn.iloc[:,4]
    x8 = uranus.iloc[:,3]; y8 = uranus.iloc[:,4]
    x9 = neptune.iloc[:,3]; y9 = neptune.iloc[:,4]
    

    ax.plot(x1,y1,label = 'Sun',linewidth=0.5)
    ax.plot(x2,y2,label = 'Mercury',linewidth=0.5,)
    ax.plot(x3,y3,label = 'Venus',linewidth=0.5)
    ax.plot(x4,y4,label = 'Earth',linewidth=0.5)
    ax.plot(x5,y5,label = 'Mars',linewidth=0.5)
#     ax.plot(x6,y6,label = r'$M_{\mathrm{J}}$ 1$\times$',linewidth=2.0)
#     ax.plot(x7,y7,label = r'$M_{\mathrm{J}}$ 1$\times$',linewidth=2.0)
#     ax.plot(x8,y8,label = r'$M_{\mathrm{J}}$ 1$\times$',linewidth=2.0)
#     ax.plot(x9,y9,label = r'$M_{\mathrm{J}}$ 1$\times$',linewidth=2.0)
    


    plt.legend(loc='upper right', frameon= False)
    
    plt.savefig("solar1.eps",facecolor='w', bbox_inches='tight',transparent=True,dpi=600)

    plt.show()
```

```python
    fig = plt.figure(figsize=(5, 5))
    ax = SubplotZero(fig, 111)
    fig.add_subplot(ax)
    ax.grid(False)
    # Major and Minor Ticks setting
    xmajorLocator   = MultipleLocator(1)
    xmajorFormatter = FormatStrFormatter('%5.1f')
    xminorLocator = MultipleLocator(0.5) 
    ax.xaxis.set_major_locator(xmajorLocator)
    ax.xaxis.set_major_formatter(xmajorFormatter)
    ax.xaxis.set_minor_locator(xminorLocator)
    ax.yaxis.set_major_locator(xmajorLocator)
    ax.yaxis.set_major_formatter(xmajorFormatter)
    ax.yaxis.set_minor_locator(xminorLocator)


    for direction in ["xzero", "yzero"]:
        ax.axis[direction].set_axisline_style("-|>")
        ax.axis[direction].set_visible(True)
        ax.axis[direction].line.set_color('k')
        ax.set_aspect(1)
       
    ax.axis["yzero"].set_ticklabel_direction("-")
    
    for direction in ["left", "right", "bottom", "top"]:
        ax.axis[direction].set_visible(False)

    Euler = pd.read_csv("Euler_5000.csv")
    Verlet = pd.read_csv("Verlet_5000.csv")
    x = Euler.iloc[:,1]; y = Euler.iloc[:,2]
    x1 = Verlet.iloc[:,1]; y1 = Verlet.iloc[:,2]
    ax.plot(x,y,label = 'Euler')
    ax.plot(x1,y1,label = 'Verlet')
    plt.legend(loc='lower right', frameon= False)
    
    plt.savefig("Earth5000.eps",facecolor='w', bbox_inches='tight',transparent=True,dpi=256)

    plt.show()
```
