```python
    import numpy as np
    import matplotlib.pyplot as plt
    import pandas as pd
    from mpl_toolkits.axes_grid.axislines import SubplotZero
    from matplotlib.ticker import MultipleLocator, FormatStrFormatter
    fig = plt.figure(figsize=(5, 5))
    ax = SubplotZero(fig, 111)
    fig.add_subplot(ax)
    ax.grid(False)
    
    # Major and Minor Ticks setting
    xmajorLocator   = MultipleLocator(1)
    xmajorFormatter = FormatStrFormatter('%d')
    xminorLocator = MultipleLocator(0.5) 
    ymajorLocator   = MultipleLocator(1)
    ymajorFormatter = FormatStrFormatter('%d')
    yminorLocator = MultipleLocator(0.5) 
    ax.xaxis.set_major_locator(xmajorLocator)
    ax.xaxis.set_major_formatter(xmajorFormatter)
    ax.xaxis.set_minor_locator(xminorLocator)
    ax.yaxis.set_major_locator(ymajorLocator)
    ax.yaxis.set_major_formatter(ymajorFormatter)
    ax.yaxis.set_minor_locator(yminorLocator)


    for direction in ["xzero", "yzero"]:
        ax.axis[direction].set_axisline_style("-|>")
        ax.axis[direction].set_visible(True)
        ax.axis[direction].line.set_color('k')
        ax.set_aspect(1)
       
    ax.axis["yzero"].set_ticklabel_direction("-")
    
    for direction in ["left", "right", "bottom", "top"]:
        ax.axis[direction].set_visible(False)

    Verlet = pd.read_csv("Verlet5_1.2.csv")
    Verlet1 = pd.read_csv("Verlet_10.csv")
    x = Verlet.iloc[:,1]; y = Verlet.iloc[:,2]
    x1 = Verlet1.iloc[:,1]; y1 = Verlet1.iloc[:,2]
    ax.plot(x,y,label = 'Verlet1.2')
    ax.plot(x1,y1,label = 'Verlet1.4')
    plt.legend(loc='lower left', frameon= False)
    
    plt.savefig("Escape.eps",facecolor='w', bbox_inches='tight',transparent=True,dpi=256)

    plt.show()
```
