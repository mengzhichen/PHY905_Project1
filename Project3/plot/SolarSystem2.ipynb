```python
    import numpy as np
    import matplotlib.pyplot as plt
    import pandas as pd
    from mpl_toolkits.axes_grid.axislines import SubplotZero
    from matplotlib.ticker import MultipleLocator, FormatStrFormatter
    fig = plt.figure(figsize=(5, 8))
    ax = SubplotZero(fig, 111)
    fig.add_subplot(ax)
    ax.grid(False)
    
    # Major and Minor Ticks setting
    xmajorLocator   = MultipleLocator(1)
    xmajorFormatter = FormatStrFormatter('%5.1f')
    xminorLocator = MultipleLocator(0.5) 
    ax.xaxis.set_major_locator(xmajorLocator)
    ax.xaxis.set_major_formatter(xmajorFormatter)
    ax.xaxis.set_minor_locator(xminorLocator)
    ax.yaxis.set_major_locator(xmajorLocator)
    ax.yaxis.set_major_formatter(xmajorFormatter)
    ax.yaxis.set_minor_locator(xminorLocator)


    for direction in ["xzero", "yzero"]:
        ax.axis[direction].set_axisline_style("-|>")
        ax.axis[direction].set_visible(True)
        ax.axis[direction].line.set_color('k')
        ax.set_ylim(-2.2, 4.4)
        ax.set_xlim(-2.2, 4.4)
        ax.set_aspect(1)
       
    ax.axis["yzero"].set_ticklabel_direction("-")
    
    for direction in ["left", "right", "bottom", "top"]:
        ax.axis[direction].set_visible(False)

    Earth1 = pd.read_csv("earth1.csv")
    Earth2 = pd.read_csv("earth2.csv")
    Earth3 = pd.read_csv("earth3.csv")
    Earth4 = pd.read_csv("earth4.csv")

 
    x = Earth1.iloc[:,3]; y = Earth1.iloc[:,4]
    x2 = Earth2.iloc[:,3]; y2 = Earth2.iloc[:,4]
    x3 = Earth3.iloc[[0,4000],3]; y3 = Earth3.iloc[[0,4000],4]
    x4 = Earth4.iloc[:,3]; y4 = Earth4.iloc[:,4]
    
#     ax.plot(x4,y4,label =  r'Jupiter mass 10000$\times$')
    ax.plot(x3,y3,label =  r'$M_{\mathrm{J}}$ 1000$\times$')
    ax.plot(x2,y2,label =  r'$M_{\mathrm{J}}$ 10$\times$',linewidth=2.0)
    ax.plot(x,y,label = r'$M_{\mathrm{J}}$ 1$\times$',linewidth=2.0)
    #ax.plot(x2,y2,label =  r'Jupiter mass 10$\times$',linewidth=1.0)
    #ax.plot(x3,y3,label =  r'Jupiter mass 1000$\times$')

    plt.legend(loc='upper right', frameon= False)
    
    #plt.savefig("Earth1.eps",facecolor='w', bbox_inches='tight',transparent=True,dpi=600)

    plt.show()
```
